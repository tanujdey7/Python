```python
x = df4.groupby(['Type 1'])[['Attack']].mean()
y = df4.groupby(['Type 2'])[['Attack']].mean()
```

```python
y
```

```python
x.reset_index(inplace=True)
y.reset_index(inplace=True)
```

```python
new= pd.merge(x,y,left_on='Type 1',right_on='Type 2')
```

```python
new.drop(['Type 1'],axis=1)
```

```python
plt.plot(new['Type 2'],new['Attack_x'],color='r')
plt.plot(new['Type 2'],new['Attack_y'],color='b')
plt.legend(['Type 1','Type 2'])
plt.show()
```
